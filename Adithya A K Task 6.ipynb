##4.  Write a Python program to plot two or more lines with legends, different widths and colors. The code snippet gives the output shown in the following screenshot:



image.png

````python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([10,20,30])
y1 = np.array([20,40,10])
x2 = np.array([10, 20, 30])
y2 = np.array([40,10,30])
plt.title("Two or more lines with different widths and colors with suitable legends")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.plot(x1,y1, color = 'blue', linewidth = '3')
plt.plot(x2,y2, color = 'red', linewidth = '5')
plt.xlim(10,30)
plt.ylim(10,40)
plt.legend(["line 1 - width - 3","line 2 - width - 5"])
plt.show()
````

##6.  Write a Python programming to display a bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["Java", "Python","PHP","JavaScript","C#","C++"])
y = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
plt.title("Popularity of Programming Language\n Worldwide, Oct 2017 compared to a year ago")
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.bar(x,y,align = 'edge',color = 'blue')
plt.grid(True,color = 'red',linestyle = '-',which = 'major',linewidth = .5)
plt.grid(True,color = 'black', linestyle = '--', which = 'minor',linewidth = .5)
plt.minorticks_on()
plt.tick_params(which = 'both')
plt.ylim(0,25)
plt.xlim("Java", " ",5)
plt.show()
````

##Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

image.png

````python
import pandas as pd
import matplotlib.pyplot as plt

dataframe = pd.read_csv("/content/drive/MyDrive/Dataset/Company_sales_data.csv")
x = dataframe ['month_number'].sum()
Data = [dataframe['facecream'].sum(),dataframe['facewash'].sum(),dataframe['toothpaste'].sum(),
        dataframe['bathingsoap'].sum(),dataframe['shampoo'].sum(),dataframe['moisturizer'].sum()]
plt.axis('equal')
plt.title("Sales Data")
plt.pie(Data,labels =['FaceCream', 'FaceWash','ToothPaste','BathingSoap','Shampoo','Moisturizer'],autopct = '%1.2f%%')
plt.legend(loc = 'lower right')
plt.show()
````

##Exercise Question 10: Read all product sales data and show it using the stack plot
The Stack plot should look like this.

image.png


````python
import pandas as pd
import matplotlib.pyplot as plt

dataframe = pd.read_csv("/content/drive/MyDrive/Dataset/Company_sales_data.csv")
x = dataframe ['month_number'].tolist()
y1 = dataframe['facecream'].tolist()
y2 = dataframe['facewash'].tolist()
y3 = dataframe['toothpaste'].tolist()
y4 = dataframe['bathingsoap'].tolist()
y5 = dataframe['shampoo'].tolist()
y6 = dataframe['moisturizer'].tolist()

plt.plot([], [], color='violet', label = 'Face Cream',linewidth = 4)
plt.plot([], [], color='blue', label = 'Face Wash',linewidth = 4)
plt.plot([], [], color='red', label = 'Tooth Paste',linewidth = 4 )
plt.plot([], [], color='black', label = 'Bathing Soap',linewidth = 4)
plt.plot([], [], color='green', label = 'Shampoo',linewidth = 4)
plt.plot([], [], color='yellow', label = 'Moisturizer',linewidth = 4)

plt.stackplot(x,y1,y2,y3,y4,y5,y6, baseline  = 'zero',colors = ['violet','blue','red','black','green','yellow'])
plt.title("All Product Sales Data Using Stack Plot")
plt.xlabel("Month Number")
plt.ylabel("Sales Units In Number")
plt.legend(loc = 'upper left')
plt.show()
````
